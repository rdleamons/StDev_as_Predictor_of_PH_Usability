```julia
include("phFun.jl")
```

```julia
# Get paths
fruitPath = "Fruits/"

# Declare empty vectors
fruitVec = [] 
```

```julia
createPD(fruitPath, fruitVec)
```

# Create Distance Matrices
### - Bottleneck

```julia
fruitBottle = pairwise(Bottleneck(), fruitVec)
```

### - Wasserstein

```julia
fruitWasser = pairwise(Wasserstein(), fruitVec)
```

# Splice Matrices
### - Banana Matrices

```julia
BananaBottle = fruitBottle[1:100, 1:100]
notBananaBottle = fruitBottle[101:500, 1:100]
```

```julia
BananaWasser = fruitWasser[1:100, 1:100]
notBananaWasser = fruitWasser[101:500, 1:100]
```

### - Blueberry Matrices

```julia
BlueberryBottle = fruitBottle[101:200, 101:200]
notBlueberryBottleA = fruitBottle[1:100, 101:200]
notBlueberryBottleB = fruitBottle[201:500, 101:200]
notBlueberryBottle = vcat(notBlueberryBottleA, notBlueberryBottleB) 
```

```julia
BlueberryWasser = fruitWasser[101:200, 101:200]
notBlueberryWasserA = fruitWasser[1:100, 101:200]
notBlueberryWasserB = fruitWasser[201:500, 101:200]
notBlueberryWasser = vcat(notBlueberryWasserA, notBlueberryWasserB) 
```

### - Cactus Fruit Matrices

```julia
CactusBottle = fruitBottle[201:300, 201:300]
notCactusBottleA = fruitBottle[1:200, 201:300]
notCactusBottleB = fruitBottle[301:500, 201:300]
notCactusBottle = vcat(notCactusBottleA, notCactusBottleB) 
```

```julia
CactusWasser = fruitWasser[201:300, 201:300]
notCactusWasserA = fruitWasser[1:200, 201:300]
notCactusWasserB = fruitWasser[301:500, 201:300]
notCactusWasser = vcat(notCactusWasserA, notCactusWasserB) 
```

### - Clementine Matrices

```julia
ClementineBottle = fruitBottle[301:400, 301:400]
notClementineBottleA = fruitBottle[1:300, 301:400]
notClementineBottleB = fruitBottle[401:500, 301:400]
notClementineBottle = vcat(notClementineBottleA, notClementineBottleB) 
```

```julia
ClementineWasser = fruitWasser[301:400, 301:400]
notClementineWasserA = fruitWasser[1:300, 301:400]
notClementineWasserB = fruitWasser[401:500, 301:400]
notClementineWasser = vcat(notClementineWasserA, notClementineWasserB) 
```

### - Strawberry Matrices

```julia
StrawberryBottle = fruitBottle[401:500, 401:500]
notStrawberryBottle = fruitBottle[1:400, 401:500]
```

```julia
StrawberryWasser = fruitWasser[401:500, 401:500]
notStrawberryWasser = fruitWasser[1:400, 401:500]
```

# Distributions and Error Calculations

```julia
P = length(BananaBottle)
N = length(notBananaBottle)
```

### - Banana: Bottleneck

```julia
minBananaThreshold = getNZMinimum(BananaBottle)
maxBananaThreshold = maximum(BananaBottle)
FP = getFPDual(minBananaThreshold, maxBananaThreshold, notBananaBottle)
babErr = dualThresholdError(P, N, FP, 0)

print("False Positives: ", FP)
print("\nError: ", dualThresholdError(P, N, FP, 0))

groupedhist(BananaBottle, seriescolor = :PRGn_4, seriesalpha = 0.8)
groupedhist!(notBananaBottle, seriescolor = :PiYG_5, seriesalpha = 0.5)
vline!([minBananaThreshold], linewidth = 3, color = :blue)
vline!([maxBananaThreshold], linewidth = 3, color = :blue)
plot!(title = "Banana vs Other Bottleneck Distances", xlabel = "Distance", ylabel = "Frequency")
```

```julia
savefig("BananaBottleneckDistribution.png")
```

```julia
minBananaThreshold = getNZMinimum(BananaWasser)
maxBananaThreshold = maximum(BananaWasser)
FP = getFPDual(minBananaThreshold, maxBananaThreshold, notBananaWasser)
bawErr = dualThresholdError(P, N, FP, 0)

print("False Positives: ", FP)
print("\nError: ", dualThresholdError(P, N, FP, 0))

groupedhist(BananaWasser, seriescolor = :PRGn_4, seriesalpha = 0.8)
groupedhist!(notBananaWasser, seriescolor = :PiYG_5, seriesalpha = 0.5)
vline!([minBananaThreshold], linewidth = 3, color = :blue)
vline!([maxBananaThreshold], linewidth = 3, color = :blue)
plot!(title = "Banana vs Other Wasserstein Distances", xlabel = "Distance", ylabel = "Frequency")
```

### Graph Standard Deviation vs Error

```julia
StDev = [babSTD, bawSTD, bbbSTD, bbwSTD, cfbSTD, cfwSTD, clbSTD, clwSTD, sbbSTD, sbwSTD; babErr, bawErr, bbbErr, bbwErr, cfbErr, cfwErr, clbErr, clwErr, sbbErr, sbwErr]
Err = [babErr, bawErr, bbbErr, bbwErr, cfbErr, cfwErr, clbErr, clwErr, sbbErr, sbwErr]
```

```julia
x = [bawSTD, bbwSTD, cfwSTD, clwSTD, sbwSTD]
y = [bawErr, bbwErr, cfwErr, clwErr, sbwErr]
annotate = ["Banana", "Blueberry", "Cactus", "Clementine", "Strawberry"]
```

```julia
plot(x, label = "Standard Deviation", color = :PRGn_4, series_annotations = text.(annotate, :bottom, 7))
plot!(y, label = "Error", color = :PiYG_5, series_annotations = text.(annotate, :bottom, 7))
plot!(title = "Wasserstein Standard Deviation vs Error", xlabel = "Fruit", ylabel = "Value")
```
